```python
import networkx as nx
import os 
```

```python
def write_graph(file_name,G):
    f=open(file_name,'w')
    for i in list(G.edges):
        f.write(str(i[0])+','+str(i[1])+'\n')
    f.close()
```

```python
KC=nx.karate_club_graph()
```

```python
nx.draw_networkx(KC)
```

```python
## storing graph in txt file for input
write_graph(os.path.join('input','karate_club_graph.txt'),KC)
```
